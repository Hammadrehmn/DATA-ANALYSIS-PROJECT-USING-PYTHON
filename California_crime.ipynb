```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Column1': [10, 20, 30, 40],
    'Column2': [50, 60, 70, 80],
    'Column3': [100, 200, 300, 400]
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Calculate percentages for each column (excluding the 'Category' column)
percentage_df = df.drop('Category', axis=1).apply(lambda x: (x / x.sum()) * 100, axis=1)

# Add the 'Category' column back to the DataFrame
percentage_df.insert(0, 'Category', df['Category'])

# Display the DataFrame with percentages
print("\nDataFrame with Percentages:")
print(percentage_df)
```
